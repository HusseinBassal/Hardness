```python
import os
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from tensorflow import keras 
from tensorflow.keras import models
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
```

```python
class MachineLearningModels:
    def __init__(self):
        pass

    def load_data(self, dataset_path):
        X = []
        y = []
    
        for label in os.listdir(dataset_path):
            label_path = os.path.join(dataset_path, label)
            if os.path.isdir(label_path):
                for file in os.listdir(label_path):
                    file_path = os.path.join(label_path, file)
                    if file.endswith(".npy"):
                        data = np.load(file_path)
                        flattened_data = data.flatten()  # Flatten the data
                        X.append(flattened_data)
                        y.append(int(label.split()[1]))  # Extract the class label from the folder name
        return np.array(X), np.array(y)

    def SVM(self, dataset_path):
        X, y = self.load_data(dataset_path)
        
        # Define number of folds for stratified split
        num_folds = 5  # Adjust as needed
        
        # Perform stratified k-fold split
        skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)
        
        avg_scores = []
        std_devs = []
        
        # Iterate over folds
        for train_index, val_index in skf.split(X, y):
            X_train, X_val = X[train_index], X[val_index]
            y_train, y_val = y[train_index], y[val_index]
            
            # Define and train your SVM model
            svm_model = SVC()
            svm_model.fit(X_train, y_train)
            
            # Evaluate the model
            avg_score = svm_model.score(X_val, y_val)
            avg_scores.append(avg_score)
        
        # Calculate average accuracy and standard deviation across folds
        avg_accuracy = np.mean(avg_scores)
        std_dev = np.std(avg_scores)
        
        return avg_accuracy, std_dev

    def KNN(self, dataset_path):
        X, y = self.load_data(dataset_path)
        
        # Define number of folds for stratified split
        num_folds = 5  # Adjust as needed
        
        # Perform stratified k-fold split
        skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)
        
        avg_scores = []
        std_devs = []
        
        # Iterate over folds
        for train_index, val_index in skf.split(X, y):
            X_train, X_val = X[train_index], X[val_index]
            y_train, y_val = y[train_index], y[val_index]
            
            # Define and train your KNN model
            knn_model = KNeighborsClassifier()
            knn_model.fit(X_train, y_train)
            
            # Evaluate the model
            avg_score = knn_model.score(X_val, y_val)
            avg_scores.append(avg_score)
        
        # Calculate average accuracy and standard deviation across folds
        avg_accuracy = np.mean(avg_scores)
        std_dev = np.std(avg_scores)
        
        return avg_accuracy, std_dev

    def SLFNN(self, dataset_path):
        X, y = self.load_data(dataset_path)
        
        # Define number of folds for stratified split
        num_folds = 5  # Adjust as needed
        
        # Perform stratified k-fold split
        skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)
        
        avg_scores = []
        std_devs = []
        
        # Iterate over folds
        for train_index, val_index in skf.split(X, y):
            X_train, X_val = X[train_index], X[val_index]
            y_train, y_val = y[train_index], y[val_index]
            
            # Define and train your SLFNN model
            slfnn_model = MLPClassifier()
            slfnn_model.fit(X_train, y_train)
            
            # Evaluate the model
            avg_score = slfnn_model.score(X_val, y_val)
            avg_scores.append(avg_score)
        
        # Calculate average accuracy and standard deviation across folds
        avg_accuracy = np.mean(avg_scores)
        std_dev = np.std(avg_scores)
        
        return avg_accuracy, std_dev
```
